```python
import numpy as np
```

```python
#   Vypracování úkolu 1

pole1 = np.array([100, 105, 98, 110, 107])

print(f"Mean: {pole1.mean()}")
print(f"Max: {pole1.max()}")
print(f"Min: {pole1.min()}")
print(f"Std: {pole1.std():.4f}")
```

```python
#   Vypracování úkolu 2

pole2 = np.array([100, 102, 101, 105, 103, 108])

#   denni procentualni vynos = ( cena dnes - cena vcera ) / cena vcera
vynos = ( pole2[1:]  -  pole2[:-1]) / pole2[:-1]
prum_vynos = vynos.mean()
celk_vynos = vynos.sum()

print(f"Vynosy: {vynos}")
print(f"Avg vynos: {prum_vynos:.4f}")
print(f"Celkovy vynos: {celk_vynos:.4f}")
```

```python
#   Vypraování úkolu 3

np.random.seed(42)
normal_vynosy = np.random.normal(0.0005, 0.02, 10000)

mask_zisk = normal_vynosy > 0
mask_ztrata = normal_vynosy < 0

kum_vynos = ((1 + normal_vynosy).cumprod()[-1] - 1)

print(f"Ziskovych dnu bylo: {mask_zisk.sum() / normal_vynosy.size * 100} %")
print(f"Prumerny zisk ziskoveho dne: {(normal_vynosy[mask_zisk].mean() * 100):.2f} %")
print(f"Prumerna ztrata ztratoveho dne: {(normal_vynosy[mask_ztrata].mean() * 100):.2f} %")
print(f"Kumulativni vynos za obdobi: {((kum_vynos) * 100):.2f} %")

```
