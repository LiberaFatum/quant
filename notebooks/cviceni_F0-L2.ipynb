```python
import numpy as np
```

```python
#VYPRACOVANI UKOLU 1:
pole = np.array([100, 105, 98, 110, 107])
print(f"Prumer: {pole.mean()}")
print(f"Minimum: {pole.min()}")
print(f"Maximum: {pole.max()}")
print(f"Směrodatná odchylka: {pole.std():.4f}")
```

```python
# VYPRACOVANI UKOLU 2:
ceny = np.array([100, 102, 101, 105, 103, 108])
vynos = ( ceny[1:] - ceny[:-1] ) / ceny[:-1]  # denni procentualni vynos
print(f"Vynosy: {vynos}")
print(f"Prumerny vynos: {vynos.mean():.4f}")
print(f"Celkovy vynos: {vynos.sum():.4f}")
```

```python
# VYPRACOVANI UKOLU 3:
np.random.seed(42)
rnd_vynos = np.random.normal(0.0005, 0.02, 10000)

maska_zisk = rnd_vynos > 0
print(f"Ziskových bylo {(maska_zisk.sum()/rnd_vynos.size*100):.4f} % dnů")

print(f"Průměrný zisk ziskového dne je: {rnd_vynos[maska_zisk].sum()/maska_zisk.sum():.4f}")    # lepe prumer pres rnd_vynos[maska_zisk].mean()

maska_ztrata = rnd_vynos < 0
print(f"Ztrátových bylo {(maska_ztrata.sum()/rnd_vynos.size*100):.4f} % dnů")

print(f"Průměrná ztráta ztrátového dne je: {rnd_vynos[maska_ztrata].sum()/maska_ztrata.sum():.4f}")    # lepe prumer pres rnd_vynos[maska_ztrata].mean()

kum_vynos = ((1 + rnd_vynos).cumprod()[-1] - 1) # bereme posledni hodnotu z pole
print(f"Kumulativni vynos za cele obdobi: {kum_vynos.sum():.4f}")
```
